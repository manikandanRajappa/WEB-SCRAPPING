```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url="https://www.flipkart.com/search?q=shoes&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_2_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_2_0_na_na_na&as-pos=2&as-type=TRENDING&suggestionId=shoes&requestId=9af3b1fb-2de4-4f4b-9c31-25635ece72c4"
```

```python
permission=requests.get(url)
```

```python
permission
```

```python
FURNITURE=BeautifulSoup(permission.text)
```

```python
FURNITURE
```

```python
PRODUCT_NAME=FURNITURE.select("._2WkVRV")
```

```python
PRODUCT_NAME
```

```python
import pandas
```

```python
product=pandas.DataFrame(PRODUCT_NAME,columns=["productname"])
```

```python
product
```

```python
product_price=FURNITURE.select("._3bPFwb ._30jeq3")
```

```python
product_price
```

```python
product["product_price"]=product_price
```

```python
product
```

```python
offer=FURNITURE.select("._3bPFwb span")
```

```python
offer
```

```python
product["offer"]=offer
```
